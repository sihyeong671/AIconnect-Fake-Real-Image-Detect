```python
from glob import glob
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_img_paths = glob("../train/*/*")
test_img_paths = glob("../test/images/*")
print(len(train_img_paths), len(test_img_paths))
```

```python
w_lst = []
h_lst = []

for path in train_img_paths:
  img = cv2.imread(path)
  h, w, _ = img.shape
  w_lst.append(w)
  h_lst.append(h)

for path in test_img_paths:
  img = cv2.imread(path)
  h, w, _ = img.shape
  w_lst.append(w)
  h_lst.append(h)

sns.scatterplot(x=w_lst, y=h_lst)
plt.show()
```
